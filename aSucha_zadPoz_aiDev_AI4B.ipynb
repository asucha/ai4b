Przeanalizujmy korelacje pomiÄ™dzy danymi cechami zbioru, pierw robiÄ…c to uÅ¼ywajÄ…c wspÃ³Å‚czynnika korelacji Pearsona. Aby to zrobiÄ‡ w celu znalezienia wystarczajÄ…co duÅ¼ej korelacji pomiÄ™dzy cechami, naleÅ¼y zagregowaÄ‡ dane odpowiednich kolumn, zapisaÄ‡ kolejno wszystkie cechy jako kolumny oraz wiersze tabeli korelacyjnej i policzyÄ‡ wartoÅ›ci wspÃ³Å‚czynnika Pearsona.

```python
import numpy as np
from scipy.stats import pearsonr

numeric_df = df.select_dtypes(include=[np.number])
columns = numeric_df.columns
n = len(columns)

# pcct - Pearson's Corr Coeff Table
pcct = pd.DataFrame(np.zeros((n, n)), index=columns, columns=columns)

# spopulowanie nowo utworzonej tablei korelacyjnej
for i, col1 in enumerate(columns):
    for j, col2 in enumerate(columns):
        corr, _ = pearsonr(df[col1], df[col2])
        pcct.iloc[i, j] = corr
```

Oraz zwizualizujmy wyÅ¼ej utworzonÄ… tabelÄ™

```python
import matplotlib.pyplot as plt
import seaborn as sns

rounded_pcct = np.round(pcct, 2)

plt.figure(figsize=(14,12))
sns.heatmap(pcct,
            annot=rounded_pcct,
            fmt='.2f',
            cmap="coolwarm",
            vmin=-1,
            vmax=1,
            annot_kws={"size": 9})

ax = plt.gca()
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Na podstawie stworzonych grafÃ³w, decydujÄ™ siÄ™ stworzyÄ‡ proste model regresyjne - Model Regresji Liniowej, oraz Klasyfikacji (Logisticznej Regresji) porÃ³wnÃ³jÄ…c cechÄ™ Zanieczyszczenia Powietrza z AlergiÄ… na PyÅ‚ki.

```python
# Model Regresji Liniowej

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

print("===== ZANIECZYSZCZENIE POWIETRZA VS ALERGIA NA PYÅKI ZAWIESZONE - REGRESJA LINIOWA =====\n")

X = df[["Air Pollution"]]
y = df["Dust Allergy"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# zeskalowanie cech
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# uczenie modelu
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# generowanie predykcji
y_pred = model.predict(X_test_scaled)

# ewaluacja modelu

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared: {r2:.4f}")
print(f"Coefficient for Air Pollution: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
```

```python
# Model Regresji Logistycznej

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

print("===== ZANIECZYSZCZENIE POWIETRZA VS ALERGIA NA PYÅKI ZAWIESZONE - REGRESJA LOGISTYCZNA =====")

y_binary = (df['Dust Allergy'] > df['Dust Allergy'].median()).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# zeskalowanie cech
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# uczenie modelu
clf = LogisticRegression(random_state=42)
clf.fit(X_train_scaled, y_train)

# generowanie predykcji
y_pred_class = clf.predict(X_test_scaled)
y_pred_prob = clf.predict_proba(X_test_scaled)[:, 1]

# ewaluacja modelu
accuracy = accuracy_score(y_test, y_pred_class)
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_class))
```
